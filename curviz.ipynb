# Water Column Current Vizualization

## Method

The essential product of ADCP measurements is a series of vectors spaced along a single water column represented the direction and magnitude of current flow in a single location. However, limitations exist on our ability to represent these vectors comprehensibly in 3D space. Therefore, these vectors may be represented as a rose plot of increasingly transparent vectors, where transparency indicates depth.

Plotting as points and tracing may show Ekman transport.

```python
#TODO create colormap at side of plot showing depth
#TODO change transparency of vectors
#TODO get source from data

import numpy as num
import matplotlib.pyplot as plot
# location of vectors (origin)
theta = num.linspace(0,2 * num.pi,8)
x = [0 for x in theta]
y = [0 for x in theta]
# x and y components of vectors
u = [num.cos(theta) for x in theta]
v = [num.sin(theta) for x in theta]


plot.quiver(x,y,u,v,scale=4)
```
